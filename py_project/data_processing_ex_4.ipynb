문제 4 : 범주형 데이터 인코딩
범주형 데이터(Categorical Data)는 문자열 형태로 표현되는 데이터를 의미합니다.

머신러닝 알고리즘은 문자열 데이터를 바로 사용할 수 없기 때문에 숫자 형태로 변환해야 합니다.
범주형 데이터를 숫자 형태로 변환하는 방법에는 여러 가지가 있는데, 대표적인 방법으로 Label Encoding과 One-Hot Encoding이 있습니다.

Label Encoding은 각 범주(Category)를 숫자로 매핑하는 방식입니다.
One-Hot Encoding은 각 범주를 이진 벡터로 표현하는 방식입니다.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# 데이터 불러오기
df = pd.read_csv('adult.csv')
```

```python
# 데이터 정보 확인
df.info()
```

```python
# 데이터 구조 확인
df.head()
```

```python
# 범주형 데이터를 Label Encoding 방법을 사용하여 인코딩
for col in ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country'] :
    encoder = LabelEncoder()
    df[col + '-encoded'] = encoder.fit_transform(df[col])
```

```python
# 결과 출력
df.filter(like='encoded')   # like : 문자열이 포함된 열만 추출
```

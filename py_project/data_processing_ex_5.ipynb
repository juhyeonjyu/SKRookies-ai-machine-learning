문제 5 : 데이터 스케일링

데이터 스케일링은 데이터의 범위를 조정하는 작업을 의미합니다.
데이터 스케일링을 통해 데이터의 분포를 조정하고, 모델의 성능을 향상시킬 수 있습니다.

대표적인 데이터 스케일링 방법으로는 StandardScaler, MinMaxScaler 등이 있습니다.
이번 문제에서는 StandardScaler를 사용하여 데이터를 표준화 하는 방법을 실습합니다.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
# 데이터 불러오기
df = pd.read_csv('winequality-red.csv')
```

```python
# 데이터 정보 확인
df.info()
```

```python
# 데이터 구조 확인
df.head()
```

```python
# 표준화
scaler = StandardScaler()
for col in ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides'] :
    df['Scaled ' + col] = scaler.fit_transform(df[[col]])
```

```python
# 결과 출력
df.filter(like='Scaled')
```
